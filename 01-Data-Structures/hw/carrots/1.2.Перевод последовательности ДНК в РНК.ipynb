```python
with open("dna.fasta.txt") as file:
     s = file.read()
```

```python
def translate_from_dna_to_rna(string):
    s = string.replace("T", "U")
    return s
```

```python
s1 = translate_from_dna_to_rna(s).split('\n')
d = {}
for line in s1:
    if line.startswith(">"):
        name = line
        d[name] = ""
    else:
        d[name] += line
d
```
